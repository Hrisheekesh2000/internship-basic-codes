```python
import numpy as np
import qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import IBMQ, Aer, transpile, assemble
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
from qiskit.extensions import Initialize
from qiskit.quantum_info import random_statevector
from math import sqrt, pi


```

```python
#quantum bit commitment
qc = QuantumCircuit(8,8)
print("Enter 0 or 1")
a=int(input())
if a==1:
    qc.h(0)
    qc.h(1)
    qc.h(2)
    qc.h(3)
    qc.h(4)
    qc.h(5)
    qc.h(6)
    qc.h(7)
qc.draw()
```

```python
list = [0, 1, 2, 3, 4, 5, 6, 7]
import random
num = random.sample(list, 4)
a=num[0]
b=num[1]
c=num[2]
d=num[3]
print(num)
```

```python
##bob is honest
qc.h(a)
qc.h(b)
qc.h(c)
qc.h(d)

#permutation sequence
import random
 
list1 = [0, 1, 2, 3, 4, 5, 6, 7]
numb = random.sample(list1 , 8)


a1=numb[0]
b1=numb[1]
c1=numb[2]
d1=numb[3]
e1=numb[4]
f1=numb[5]
g1=numb[6]
h1=numb[7]

qc.cx([a1,b1],[b1,a1])
qc.cx([c1,d1],[d1,c1])
qc.cx([e1,f1],[f1,e1])
qc.cx([g1,h1],[h1,g1])

qc.barrier()

list1 = [0, 1, 2, 3, 4, 5, 6, 7]
numbr = random.sample(list1 , 4)
a2=numbr[0]
b2=numbr[1]
c2=numbr[2]
d2=numbr[3]
print(numbr)
qc.h(a2)
qc.h(b2)
qc.h(c2)
qc.h(d2)

qc.measure(list,list)
qc.draw()

```

```python
sim = Aer.get_backend('aer_simulator')
t_qc = transpile(qc, sim)
t_qc.save_statevector()
counts = sim.run(t_qc).result().get_counts()

print(counts)
```
